استفاده کنیم Train/Test برای سنجش اینکه آیا مدل به اندازه کافی خوب است، می‌توانیم از روشی به نام

مجموعه داده‌ها را به دو مجموعه تقسیم می‌کنید: یک مجموعه آموزشی و یک مجموعه آزمایشی

```python
import numpy
import matplotlib.pyplot as plt
```

### با یک مجموعه داده شروع میکنیم

```python
numpy.random.seed(5)
x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

## تقسیم به آموزش/آزمون

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

### نمایش مجموعه آزمایش و آموزش

```python
plt.scatter(train_x, train_y, color= "Blue")
plt.scatter(test_x, test_y, color= "red")
plt.show()
```

### یک خط رگرسیون چندجمله‌ای را از طریق نقاط داده رسم کنید

```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

این نتیجه می‌تواند پیشنهاد من در مورد برازش مجموعه داده‌ها با رگرسیون چندجمله‌ای را تأیید کند

اگرچه اگر سعی کنیم مقادیر خارج از مجموعه داده‌ها را پیش‌بینی کنیم، نتایج عجیبی به ما می‌دهد

## R²

هم شناخته می‌شود R2 که با نام R-squared


را اندازه‌گیری می‌کند و مقدار آن از ۰ تا ۱ متغیر است y و محور  x این شاخص رابطه بین محور


که ۰ به معنای عدم وجود رابطه و ۱ به معنای کاملاً مرتبط است

```python
from sklearn.metrics import r2_score

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```

نتیجه ۰.۷۹۹ نشان می‌دهد که رابطه‌ی قابل قبولی وجود دارد
___

حالا ما مدلی ساخته‌ایم که حداقل از نظر داده‌های آموزشی قابل قبول است

حالا می‌خواهیم مدل را با داده‌های آزمایشی نیز آزمایش کنیم تا ببینیم آیا نتیجه مشابهی به ما می‌دهد یا خیر

```python
r2 = r2_score(test_y, mymodel(test_x))

print(r2)
```

 نتیجه ۰.۸۰۹ نشان می‌دهد که مدل با مجموعه تست نیز برازش دارد و ما مطمئن هستیم که می‌توانیم از این مدل برای پیش‌بینی مقادیر آینده استفاده کنیم

حالا که مطمئن شدیم مدل ما درست کار می‌کند، می‌توانیم شروع به پیش‌بینی مقادیر جدید کنیم

## پیش‌بینی‌
